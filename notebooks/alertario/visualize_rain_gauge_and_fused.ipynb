```python
print("Time mismatch example!!!")

import pandas as pd
import numpy as np

index_15min = pd.date_range(start='2024-04-01', end='2024-04-05', freq='15T')
data_15min = {'temperature': np.random.randint(10, 30, len(index_15min))}
df_15min = pd.DataFrame(data_15min, index=index_15min)

index_1hour = pd.date_range(start='2024-04-01', end='2024-04-05', freq='1H')
data_1hour = {'pressure': np.random.randint(980, 1020, len(index_1hour))}
df_1hour = pd.DataFrame(data_1hour, index=index_1hour)

df_merged = pd.merge(df_15min, df_1hour, how='left', left_index=True, right_index=True)
df_merged.head(10)
```
