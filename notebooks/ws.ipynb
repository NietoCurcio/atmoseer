```python
import os
import pandas as pd

def read_parquet_files(dir_path):
    # Initialize an empty list to store DataFrames
    dfs = []

    # Iterate over all files and directories in dir_path
    for entry in os.scandir(dir_path):
        print(f"Current entry: {entry}")
        if entry.is_file() and entry.name.endswith('.parquet'):
            # If the entry is a CSV file, read it into a DataFrame and append it to the list
            print(f"Reading file {entry.path}")
            try:
                df = pd.read_parquet(entry.path)
                dfs.append(df)
            except pd.errors.EmptyDataError:
                print(f"File {entry.path} is empty, skipping.")
            except pd.errors.ParserError:
                print(f"Error parsing file {entry.path}, skipping.")
        elif entry.is_dir():
            # If the entry is a directory, recursively call the function on the subdirectory
            sub_dir_path = os.path.join(dir_path, entry.name)
            print(f"Entering directory {sub_dir_path}")
            dfs.extend(read_parquet_files(sub_dir_path))

    # Concatenate all DataFrames in the list into a single DataFrame and return it
    return pd.concat(dfs, ignore_index=True)

# Example usage: recursively read all CSV files in the directory "data"
data_dir = "../data/ws/alertario"
df = read_parquet_files(data_dir)

# Do something with the concatenated DataFrame, e.g., print the first 5 rows
print(df.head())
```
