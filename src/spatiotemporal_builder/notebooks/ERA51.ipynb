```python
import folium
```

```python
region_of_interest = {'north': -22, 'west': -44, 'south': -23, 'east': -42}

center_lat = (region_of_interest['north'] + region_of_interest['south']) / 2
center_lng = (region_of_interest['west'] + region_of_interest['east']) / 2

m = folium.Map(location=[center_lat, center_lng], zoom_start=9, tiles='cartodbpositron')

folium.Rectangle(
    bounds=[
        [region_of_interest['south'], region_of_interest['west']],
        [region_of_interest['north'], region_of_interest['east']]
    ],
    color='blue',
    fill=True,
    fill_opacity=0.2
).add_to(m)

m
```
